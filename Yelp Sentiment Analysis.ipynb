# Yelp Reviews Sentiment Prediction
https://www.kaggle.com/datasets/marklvl/sentiment-labelled-sentences-data-set

```python
import numpy as np
import pandas as pd
```

```python
with open('yelp.txt', 'r') as file:
    lines = file.read()
lines
```

```python
lines = lines.split('\n')
lines
```

```python
X_data = []
y_data = []
for i in range(len(lines)):
    if lines[i]:
        X_data.append(lines[i][:-3])
        y_data.append(int(lines[i][-1]))
    else:
        continue
    
y_data
```

```python
# Use try except to avoid error due to this unwanted entry
lines[1000]
```

```python
X_data = np.array(X_data)
y_data = np.array(y_data)
```

```python
X_data.shape, y_data.shape
```

```python
df = pd.DataFrame({
    'X': X_data,
    'y': y_data
})
```

```python
df
```

```python
df.y.nunique()
```

```python
df.y.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_count = v.fit_transform(X_data)
X_count.toarray()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB, MultinomialNB

from sklearn.model_selection import GridSearchCV
```

```python
model_params = {
    'logistic_regression': {
        'model': LogisticRegression(max_iter=10000, random_state=3),
        'params': {
            'C': [1, 2, 4, 8, 16, 32]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(random_state=3),
        'params': {
            'criterion': ['gini', 'entropy']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(random_state=3),
        'params': {
            'n_estimators': [100, 200, 300, 400],
            'criterion': ['gini', 'entropy', 'log_loss']
        }
    },
    'svm': {
        'model': SVC(gamma='auto', random_state=3),
        'params': {
            'C': [1, 2, 4, 8, 16, 32],
            'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
        }
    },
    'bernoulli_nb': {
        'model': BernoulliNB(),
        'params': {
            'alpha': [0.00001, 0.0001, 0.001, 0.1, 1, 10, 100,1000]
        }
    },
    'multinomial_nb': {
        'model': MultinomialNB(),
        'params': {
            'alpha': [0.00001, 0.0001, 0.001, 0.1, 1, 10, 100,1000]
        }
    }
}
```

```python
scores = []

for name, models in model_params.items():
    clf = GridSearchCV(models['model'], models['params'], cv=5, return_train_score=False)
    clf.fit(X_count, y_data)
    scores.append({
        'model': name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    print(name + ' 100%')

scores_df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
scores_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_count, y_data, test_size=0.3)

model = RandomForestClassifier(criterion='entropy', n_estimators=200)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
sentence = ['food is not good']

sentence_count = v.transform(sentence)
sentence_count.toarray()
```
